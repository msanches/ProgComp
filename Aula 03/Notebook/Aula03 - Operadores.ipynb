# 3- Módulo matemático

```python
# Primeiro devemos importar o módulo
import math
```

```python
# math.fabs(x) - Retorna valor absoluto
x = int(input('Digite um número negativo: '))
print(math.fabs(x))
```

```python
# math.floor(x) - Retorna o maior número inteiro menor ou igual a x
x = float(input('Digite um número decimal: '))
print(math.floor(x))
```

```python
# math.ceil(x) - Retorna o menor número inteiro maior ou igual a x
x = float(input('Digite um número decimal: '))
print(math.ceil(x))
```

```python
# math.sqrt(x) - Retorna a raiz quadrada de x
x = float(input('Digite um número: '))
print(math.sqrt(x))
```

```python
# math.trunc(x) - Retorna a parte inteira de x
x = float(input('Digite um número decimal: '))
print(math.trunc(x))
```

```python
# math.factoria(x) - Retorna o fatorial de x
x = int(input('Digite um número inteiro '))
print(math.factorial(x))
```

```python
# Funções trigonométricas
x = float(input('Digite um ângulo em graus: '))
x = math.radians(x) # converte o valor digitado em radianos
print(math.sin(x)) # imprime o seno de x
print(math.cos(x)) # imprime o cosseno de x
print(math.tan(x)) # imprime a tangente de x
```
